```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

data = sns.load_dataset("tips")
data
sns.scatterplot(data=data,x="total_bill",y="tip")
plt.show()
```

```python
sns.scatterplot(data=data,x="total_bill",y="tip",hue="day",size="size",palette='viridis')
plt.show()
```

```python
data=pd.read_excel("ESD.xlsx")

sns.scatterplot(data=data,x="Age",y="Annual Salary",size="Bonus %",hue="Gender")
plt.legend(bbox_to_anchor=(.7,1.2))
plt.show()
```

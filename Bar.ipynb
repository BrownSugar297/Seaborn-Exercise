```python
import seaborn as sns
import matplotlib.pyplot as plt

data=sns.load_dataset("tips")
data
```

```python
sns.barplot(data=data,x='day',y='tip')
plt.show()
```

```python
sns.barplot(data=data,x='day',y='tip',estimator='median',palette="spring")
plt.show()
```

```python
sns.barplot(data=data,x='day',y='tip',estimator='sum',hue='sex')
plt.show()
```

```python
sns.barplot(data=data,x='day',y='tip',hue='sex',palette="spring",order=["Sun","Sat","Fri","Thur"],errorbar=("ci",0))
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

Ashik=sns.load_dataset("tips")

sns.countplot(data=Ashik,x="sex")
plt.show()
```

```python
sns.countplot(data=Ashik,x="day",palette="spring")
plt.show()
```

```python
df=pd.read_excel("ESD.xlsx")

sns.countplot(data=df,y="Department",hue="Gender")
plt.show
```
